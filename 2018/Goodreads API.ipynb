# Goodreads API

```python
import requests, xmltodict, json, pprint

endpoint = "https://www.goodreads.com/search.xml?key=WI4Ik3vNIOss4Il585hw&q=Ender%27s+Game"
req = requests.request('GET', endpoint)


content = json.loads(json.dumps(xmltodict.parse(req.content)))
pp = pprint.PrettyPrinter()
pp.pprint(content)
```

```python
pp.pprint(content['GoodreadsResponse']['search']['results']['work'][0])
```
