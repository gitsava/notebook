# Strings


Komputer menyimpan karakternya dalam bentuk angka, tiap karakter yang digunakan dipasangkan dengan suatu angka yang unik.

Beberapa karakter ini ada yang disebut __whitespaces__ (misalnya spasi dan tab), dan __control characters__ yang digunakan untuk mengatur input dan output

ASCII (American Standard Code for Information Iterchange) adalah kumpulan kode yang paling banyak digunakan di dunia komputer. setiap karakter di ASCII dipasangkan dengan suatu karakter, misalnya huruf __A__ koresponden dengan __65__..



## I18N

Software I18N atau INTERNATIONALIZATION (Internasionalisasi) yang dimaksud adalah suatu program/softwae yang ditulis harapannya bisa digunakan di seluruh dunia internasional, dengan berbagai budaya, agama, dan alfabet yang digunakan



### Code points and code pages

Code point adalah nilai yang dipasangkan dengan suatu karakter, contohnya tadi 65 adalah A, 66 adalah B, dst. pada ASCII 32 code point pertama (0-31) yang digunakan untuk control characters, sisanya untuk karakter yang biasa kita gunakan. Disini dapat dikatakan kalau ASCII memiliki 128 code point.

Code page sendiri adalah standar yang menggunakan code point diatas 128 yang digunakan untuk karakter-karakter unik untuk negara nya masing - masing 

Misalnya, Code point 200 pada code Page __ISO/IEC 8859-2__ adalah __Č__ (karakter slavia(slavic) di Eropa Timur). 

Sedangkan Code point 200 pada code Page __ISO/IEC 8859-5__ adalah __Ш__ (karakter kiril (cyrillic))

Sedangkan Code point 200 pada code Page __ISO/IEC 8859-6__ adalah __ب__ (karakter arab)


## Unicode

Code page sendiri sudah membantu industri komputer dan teknologi informasi untuk menyelesaikan I18N untuk beberapa waktu, tetapi setelah diteliti lebih lanjut, ini bukanlah solusi yang permanen.

Unicode meng-assign masing-masing karakter (huruf, tanda baca, simbol, dll) menjadi jutaan code points. 128 code point pertama dari unicode sama dengan ASCII, dan 256 code point pertama dari unicode sama dengan code page ISO/IEC 8859-1(code page untuk Eropa Barat, code page yang paling sering digunakan pada saat itu)

### UTF-8

Unicode sendiri walaupun sudah menyelesaikan permasalahn I18N, Unicode menimbulkan masalah baru yakni ukurannya dan jumlahnya yang besar. Maka dari itu dibuatlah UTF-8, yang menggunakan prinsip UTF tetapi menggunakan memori lebih kecil.

UTF sendiri merupakan singkatan dari Unicode Transformation Format. UTF-8 menggunakan jumlah bit yang cukup untuk tiap code point yang direpresentasikan oleh karakter itu.

Misalnya:
- semua huruf latin (dan semua karakter standar ASCII) memakai 8 bits
- huruf non-latin memakai 16 bits
- huruf CJK (China-Japan-Korea) menmakai 24 bits

Python3 sendiri sudah menggunakan unicode dan UTF-8, yang menandakan bawah python3 sudah I18N

## Caesar Chiper 

```python
# Caesar cipher
text = input("Enter your message: ")

def doCipher(text):
    cipher = ''
    for char in text:
        if not char.isalpha():
            continue
        char = char.upper()
        code = ord(char) + 1
        if code > ord('Z'):
            code = ord('A')
        cipher += chr(code)
    return cipher

def deCipher(text):
    cipher = ''
    for char in text:
        if not char.isalpha():
            continue
        char = char.upper()
        code = ord(char) - 1
        if code < ord('A'):
            code = ord('Z')
        cipher += chr(code)
    return cipher

print(deCipher(doCipher(text)))
```

Jika kita mempunyaki kode-kode "mencurigakan" yang mungkin dapat menimpilkan eksepsi,kita perlu menyimpan kode tersebut pada blok `try`. 

Ketika kode berjalan normal, kode pada bagian `except` tidak akan dieksekusi. Sebaliknya jika terjadi kesalahan, maka eksekusi kode di bagian `try` akan dihentikan, dan program akan mengeksekusi pada bagian `except`.

```python
try:
    firstNumber = int(input("Enter the first number: "))
    secondNumber = int(input("Enter the second number: "))
    print(firstNumber / secondNumber)
except:
    print("This operation cannot be done.")

print("THE END.")
```

```python
try:
    print("1")
    x = 1 / 0
    print("2")
except:
    print("Oh dear, something went wrong...")

print("3")
```

Block `try ...  except` dapat digunakan untuk menangani lebih dari satu eksepsi, dengan menggunkan bentuk umum: 

`
try:
    :
except exc1:
    :
except exc2:
    :
except:
    :`
    
Satu atau beberapa statement yang terdapat dalam blok `try` dapat menimbulkan lebih dari satu tipe ekspesi. 

```python
try:
    x = int(input("Enter a number: "))
    y = 1 / x
    print(y)
except ZeroDivisionError:
    print("You cannot divide by zero, sorry.")
except ValueError:
    print("You must enter an integer value.")
except:
    print("Oh dear, something went wrong...")

print("THE END.")
```

Percobaan:
- Inputkan bilangan integer (e.g., 5)
- Inputkan 0
- Inputkan data non-integer


## Built-in Exception

Python 3 memnyediakan  __63 built-in exceptions__ dan semuanya memiliki hierarki seperti pohon dibawah ini:




- `ZeroDivisionError` merupakan spesial kasus dari kelas eksepsi ArithmeticError;
- `ArithmeticError` merupakan spesial kasus dari kelas eksepsi Exception;
- `Exception` merupakan spesial kasus dari kelas eksepsi BaseException;

lengkapnya bisa dilihat dalam tree berikut ini : https://i.imgur.com/D7INWGI.png

```python
try:
    y = 1 / 0
except ZeroDivisionError:
    print("Oooppsss...")
    
try:
    y = 1 / 0
except ArithmeticError:
    print("Oooppsss...")

try:
    y = 1 / 0
except Exception:
    print("Oooppsss...")
    
try:
    y = 1 / 0
except BaseException:
    print("Oooppsss...")
```

Jika eksepsi muncul di dalam fungsi, maka eksepsi itu dapat ditangani dengan dua cara:
- di dalam fungsi
- di luar fungsi

```python
def badFun(n):
    try:
        return 1 / n
    except ArithmeticError:
        print("Arithmetic Problem!")
    return None

badFun(0)

print("THE END.")
```

```python
def badFun(n):
    return 1 / n

try:
    badFun(0)
except ArithmeticError:
    print("What happened? An exception was raised!")

print("THE END.")
```

## `Raise`

Eksepsi tertentu dapat kita panggil secara pakasa dengan menggunakan perintah `raise`, meskipun sebenarnya ada kejadian yang menyebabkan jenis kesalahan tersebut

```python
def badFun(n):
    raise ZeroDivisionError
```

`raise` digunakan untuk memanggil secara paksa dari suatu eksepsi, meskipun tidak ada kejadian yang menyebabkan jenis kesalahan tersebut.

```python
def badFun(n):
    1/n
    raise ZeroDivisionError

try:
    badFun(0)
except ArithmeticError:
    print("What happened? An error?")

print("THE END.")
```

```python
def badFun(n):
    try:
        return n / 0
    except:
        print("I did it again!")
        raise

try:
    badFun(1)
except ArithmeticError:
    print("I see!")

print("THE END.")
```

Pemanggilan `raise` tanpa nama eksepsi hanya dapat dilakukan di dalam bagian `except`

Dari kode di atas `ZeroDivisionError` muncul sebanyak dua kali, yaitu:
- di dalam `try`
- di bagian`except` di dalam fungsi


## Assert

Fungsi assertion:

- Assert akan mengevaluasi ekspresi
- Jika ekspresi bernilai True atau nilai numerik bukan nol, atau string tidak kosong, atau nilai lain yang berbeda dari `None` tidak akan di eksekusi
- Jika selain itu akan muncul eksepsi `AssertionError`

Penggunaan assertion:
- kita dapat menggunakan assertion jika kita ingin kode yang dibuat benar-benar aman dari data kita belum yakin kebenarannya
- mengamankan kode dari hasil yang tidak valid
- assertion merupakan pelengkap `exception`

```python
import math

x = float(input("Enter a number: "))
assert x >= 0.0

x = math.sqrt(x)

print(x)
```
